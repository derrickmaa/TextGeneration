```python
'''制作csv（包含label)，名词和名词组合起来,指代词的消除(まだ)'''
import csv
import MeCab
from collections import Counter


t = MeCab.Tagger()
with open('C:/Users/mmmm/Desktop/cleandatanew2.txt', 'r', encoding='utf-8') as r:
    lines=r.readlines()
with open('C:/Users/mmmm/Desktop/new2.csv', 'w', encoding='utf-8', newline='') as w:
    writer = csv.writer(w)
    csv_head = ["ID","title","subtitle","paragraph","text","titlelabels","subtitlelabels","textlabels"]
    writer.writerow(csv_head)    
    n = 1
    para = 1
    for i in range(1000):
        data =[]
        if '****' in lines[i]:
            a = []
            temtitle = lines[i][4:-1]
            p = t.parseToNode(temtitle)
            tema ='no'
            while p:
                if '名詞' in p.feature and '数' not in p.feature and '人名' not in p.feature:
                    if tema == 'yes':
                        a[-1] = a[-1] + p.surface
                    else:
                        a.append(p.surface)
                    tema = 'yes'
                else:
                    tema = 'no'
                p =p.next
            temsubtitle = 'NO'            
            #print(temtitle)
            para = 1
        elif '@@' in lines[i]:
            b = []
            temsubtitle = lines[i][2:-1]
            p = t.parseToNode(temsubtitle)
            temb = 'no'
            while p:
                if '名詞' in p.feature and '数' not in p.feature and '人名' not in p.feature:
                    if temb == 'yes':
                        b[-1] = b[-1] + p.surface
                    else:
                        b.append(p.surface)
                    temb = 'yes'
                else:
                    temb = 'no'
                p = p.next
        else:
            c = []
            p = t.parseToNode(lines[i])
            temc = 'no'
            while p:
                if '名詞' in p.feature and '数' not in p.feature and '人名' not in p.feature:
                    if temc == 'yes':
                        c[-1] = c[-1] + p.surface
                    else:
                        c.append(p.surface)
                    temc = 'yes'
                else:
                    temc = 'no'
                p = p.next
            #print(Counter(c).most_common(5))
            d = Counter(c).most_common(5)
            data = [n, temtitle, temsubtitle, para, lines[i][:-1], a, b, d]
            para += 1
            n += 1
            writer.writerow(data)
```
