```python
'''段落字数统计'''
with open('C:/Users/derri/Desktop/cleandatanew.txt', 'r', encoding='utf-8') as r:
    lines = r.readlines()
    a=b=c=d=e=f=g=h=i=j=k=l=m=n=o=p=q=r=s=t=u=0
    num = 0
    for line in lines: 
        if '****' not in line and '@@' not in line:
            num += 1
            if len(line) <= 100:
                a += 1
            elif 100 < len(line) <= 200:
                b += 1
            elif 200 < len(line) <= 300:
                c += 1
            elif 300 < len(line) <= 400:
                d += 1
            elif 400 < len(line) <= 500:
                e += 1
            elif 500 < len(line) <= 600:
                f += 1
            elif 600 < len(line) <= 700:
                g += 1
            elif 700 < len(line) <= 800:
                h += 1
            elif 800 < len(line) <= 900:
                i += 1
            elif 900 < len(line) <= 1000:
                j += 1
            elif 1000 < len(line) <= 1100:
                k += 1
            elif 1100 < len(line) <= 1200:
                l += 1
            elif 1200 < len(line) <= 1300:
                m += 1
            elif 1300 < len(line) <= 1400:
                n += 1
            elif 1400 < len(line) <= 1500:
                o += 1
            elif 1500 < len(line) <= 1600:
                p += 1
            elif 1600 < len(line) <= 1700:
                q += 1
            elif 1700 < len(line) <= 1800:
                r += 1
            elif 1800 < len(line) <= 1900:
                s += 1
            elif 1900 < len(line) <= 2000:
                t += 1
            elif len(line) >= 2000:
                u += 1
        else:
            continue
    print("キャラクタは100以内：%d, 全部の %.2f" %(a, (a/num)*100))
    print("キャラクタは100-200：%d, 全部の　%.2f" %(b, (b/num)*100))
    print("キャラクタは200-200：%d, 全部の　%.2f" %(c, (c/num)*100))
    print("キャラクタは300-200：%d, 全部の　%.2f" %(d, (d/num)*100))
    print("キャラクタは400-200：%d, 全部の　%.2f" %(e, (e/num)*100))
    print("キャラクタは500-200：%d, 全部の　%.2f" %(f, (f/num)*100))
    print("キャラクタは600-200：%d, 全部の　%.2f" %(g, (g/num)*100))
    print("キャラクタは700-200：%d, 全部の　%.2f" %(h, (h/num)*100))
    print("キャラクタは800-200：%d, 全部の　%.2f" %(i, (i/num)*100))
    print("キャラクタは900-200：%d, 全部の　%.2f" %(j, (j/num)*100))
    print("キャラクタは1000-200：%d, 全部の　%.2f" %(k, (k/num)*100))
    print("キャラクタは1100-200：%d, 全部の　%.2f" %(l, (l/num)*100))
    print("キャラクタは1200-200：%d, 全部の　%.2f" %(m, (m/num)*100))
    print("キャラクタは1300-200：%d, 全部の　%.2f" %(n, (n/num)*100))
    print("キャラクタは1400-200：%d, 全部の　%.2f" %(o, (o/num)*100))
    print("キャラクタは1500-200：%d, 全部の　%.2f" %(p, (p/num)*100))
    print("キャラクタは1600-200：%d, 全部の　%.2f" %(q, (q/num)*100))
    print("キャラクタは1700-200：%d, 全部の　%.2f" %(r, (r/num)*100))
    print("キャラクタは1800-200：%d, 全部の　%.2f" %(s, (s/num)*100))
    print("キャラクタは1900-200：%d, 全部の　%.2f" %(t, (t/num)*100))
    print("キャラクタは2000以上：%d, 全部の　%.2f" %(u, (u/num)*100))
```
